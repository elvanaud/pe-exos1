```python
import numpy as np
import matplotlib.pyplot as plt
print(np.__version__)
```

```python
squares = np.full((10, 10), True)
plt.imshow(squares);
```

```python
def cadre(tab,*,inPlace = True):
    if not inPlace: tab = np.copy(tab)
    tab[1,:] = 0 #sets to 0 the first and second to last rows
    tab[-2,:] = 0
    tab[:,1] = 0 #same for the columns
    tab[:,-2] = 0
    return tab #returns the array, useful when inPlace is False
```

```python
print("Before calling 'cadre':\n", squares)
cadre(squares, inPlace = True)
plt.imshow(squares);
print("After calling 'cadre':\n", squares)
```

It is interesting to note that matplotlib changes the colorscheme in between calls to imshow wich is confusing   
We can confirm that the function cadre behaves correctly by looking at values printed in the console

```python
random = np.random.randint(0,21, dtype = np.uint8, size = (6,8))
plt.imshow(random);
```

```python
cadre(random)
plt.imshow(random);
```

The function `cadre` seems to be working as we can see the expected dark outline

Function `cadre2` adds a frame *around* the array by extending it, so in place modification isn't possible

```python
def cadre2(tab):
    h,w = tab.shape
    newTab = np.zeros(shape = (h+2,w+2), dtype = tab.dtype) #not efficient to fill an entire array withs zeroes to overwrite the
    newTab[1:-1,1:-1] = tab
    return newTab
```

```python
framedRandom = cadre2(random)
plt.imshow(framedRandom);
```
